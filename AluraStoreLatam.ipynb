### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

#1. Análisis de facturación



```python
Ingreso_total = tienda['Precio'].sum()
Ingreso_total2 = tienda2['Precio'].sum()
Ingreso_total3 = tienda3['Precio'].sum()
Ingreso_total4 = tienda4['Precio'].sum()
```

```python
print(f'El ingreso total de la tienda 1 es: ${Ingreso_total}')
print(f'El ingreso total de la tienda 2 es: ${Ingreso_total2}')
print(f'El ingreso total de la tienda 3 es: ${Ingreso_total3}')
print(f'El ingreso total de la tienda 4 es: ${Ingreso_total4}')
```

```python
# Creo una serie para los ingresos totales por tienda
ingresos_totales = pd.Series({
    'Tienda 1': Ingreso_total,
    'Tienda 2': Ingreso_total2,
    'Tienda 3': Ingreso_total3,
    'Tienda 4': Ingreso_total4
})

# Defino los colores para la reproducción gráfica
store_colors = {
    'Tienda 1': 'skyblue',
    'Tienda 2': 'lightcoral',
    'Tienda 3': 'lightgreen',
    'Tienda 4': 'gold'
}

# Creo el gráfico de barras
plt.figure(figsize=(10, 6))
ingresos_totales.plot(kind='bar', color=[store_colors[store] for store in ingresos_totales.index])
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingreso Total ($)')
plt.ticklabel_format(style='plain', axis='y')
plt.ylim(1000000000, 1200000000) # Establece el rango del eje y para mejorar visualización
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Gráfico de torta para los ingresos totales por tienda
plt.figure(figsize=(10, 8))
ingresos_totales.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=[store_colors[store] for store in ingresos_totales.index])
plt.title('Distribución de Ingresos Totales por Tienda')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

**Ingresos totales**

El análisis de ingresos muestra diferencias relevantes entre las tiendas. La **tienda 4** presenta un nivel de facturación **considerablemente inferior en comparación con las demás.**

Esto sugiere:

- menor volumen de ventas
- menor rotación de productos
- menor contribución al resultado global del negocio

# 2. Ventas por categoría

```python
cantidadesporcategoria = tienda.groupby('Categoría del Producto').size().sort_values(ascending=False)
cantidadesporcategoria2 = tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False)
cantidadesporcategoria3 = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False)
cantidadesporcategoria4 = tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False)
```

```python
print("Cantidad de productos vendidos por categoría - Tienda 1")
print(cantidadesporcategoria)
print("Cantidad de productos vendidos por categoría - Tienda 2")
print(cantidadesporcategoria2)
print("Cantidad de productos vendidos por categoría - Tienda 3")
print(cantidadesporcategoria3)
print("Cantidad de productos vendidos por categoría - Tienda 4")
print(cantidadesporcategoria4)
```

```python
ventasporcategoria = tienda.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)

ventasporcategoria2 = tienda2.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)

ventasporcategoria3 = tienda3.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)

ventasporcategoria4 = tienda4.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
```

```python
print("Ganacia total según productos vendidos por categoría - Tienda 1")
print(ventasporcategoria)
print("Ganacia total según productos vendidos por categoría - Tienda 2")
print(ventasporcategoria2)
print("Ganacia total según productos vendidos por categoría - Tienda 3")
print(ventasporcategoria3)
print("Ganacia total según productos vendidos por categoría - Tienda 4")
print(ventasporcategoria4)
```

```python
# Combino la serie de ventasporcategoria en solo df
ventas_por_categoria_df = pd.DataFrame({
    'Tienda 1': ventasporcategoria,
    'Tienda 2': ventasporcategoria2,
    'Tienda 3': ventasporcategoria3,
    'Tienda 4': ventasporcategoria4
})

# Ordeno el df por las suma de las ventas de cada tienda
ventas_por_categoria_df = ventas_por_categoria_df.sum(axis=1).sort_values(ascending=False).index

ventasporcategoria_reordered = ventasporcategoria.loc[ventas_por_categoria_df]
ventasporcategoria2_reordered = ventasporcategoria2.loc[ventas_por_categoria_df]
ventasporcategoria3_reordered = ventasporcategoria3.loc[ventas_por_categoria_df]
ventasporcategoria4_reordered = ventasporcategoria4.loc[ventas_por_categoria_df]

plt.figure(figsize=(16, 9))

# Grafico las venta por categoria de cada tienda
plt.plot(ventasporcategoria_reordered.index, ventasporcategoria_reordered.values, marker='o', linestyle='-', label='Tienda 1', color=store_colors['Tienda 1'])
plt.plot(ventasporcategoria2_reordered.index, ventasporcategoria2_reordered.values, marker='o', linestyle='-', label='Tienda 2', color=store_colors['Tienda 2'])
plt.plot(ventasporcategoria3_reordered.index, ventasporcategoria3_reordered.values, marker='o', linestyle='-', label='Tienda 3', color=store_colors['Tienda 3'])
plt.plot(ventasporcategoria4_reordered.index, ventasporcategoria4_reordered.values, marker='o', linestyle='-', label='Tienda 4', color=store_colors['Tienda 4'])

plt.title('Comparación de Ventas por Categoría de Producto entre Tiendas')
plt.xlabel('Categoría del Producto')
plt.ylabel('Ganancia Total ($)')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

**Tiendas que destacan (mejor desempeño)
Tienda 1 y Tienda 2**

Estas tiendas destacan porque:
- Tienen mayor volumen de ventas en varias categorías.
- Presentan mayor diversificación de productos vendidos.
- Generan más ingresos en categorías de alto valor como:
-- Electrónicos
-- Electrodomésticos
-- Muebles

Esto indica un portafolio más equilibrado y dinámico, lo que normalmente se traduce en mayor facturación total.

**Tienda con desempeño intermedio
Tienda 3**

- Tiene ventas relativamente estables.
- Participa en varias categorías.

Sin embargo, no alcanza el volumen de ventas de las tiendas 1 y 2.

Se puede considerar una tienda funcional pero no líder.

**Tienda con menor desempeño
Tienda 4**

Esta tienda presenta:
- Menor volumen total de ventas
- Mayor concentración en pocas categorías
- Menor dinamismo comercial

Esto significa que:
- Vende menos variedad de productos
- Genera menos ingresos por categoría
- Tiene menor impacto en los ingresos globales

# 3. Calificación promedio de la tienda


```python
calificacion_promedio = tienda['Calificación'].mean()
calificacion_promedio2 = tienda2['Calificación'].mean()
calificacion_promedio3 = tienda3['Calificación'].mean()
calificacion_promedio4 = tienda4['Calificación'].mean()
```

```python
print(f'La calificación promedio de la tienda 1 es: {calificacion_promedio:.2f}')
print(f'La calificación promedio de la tienda 2 es: {calificacion_promedio2:.2f}')
print(f'La calificación promedio de la tienda 3 es: {calificacion_promedio3:.2f}')
print(f'La calificación promedio de la tienda 4 es: {calificacion_promedio4:.2f}')
```

```python
calificaciones_promedio = pd.Series({
    'Tienda 1': calificacion_promedio,
    'Tienda 2': calificacion_promedio2,
    'Tienda 3': calificacion_promedio3,
    'Tienda 4': calificacion_promedio4
})


plt.figure(figsize=(10, 6))
calificaciones_promedio.plot(kind='bar', color=[store_colors[store] for store in calificaciones_promedio.index])
plt.title('Calificación Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Calificación Promedio')
plt.ylim(3.95, 4.06)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
envio_promedio = tienda['Costo de envío'].mean()
envio_promedio2 = tienda2['Costo de envío'].mean()
envio_promedio3 = tienda3['Costo de envío'].mean()
envio_promedio4 = tienda4['Costo de envío'].mean()
```

```python
print(f'El costo de envío promedio de la tienda 1 es: ${envio_promedio:.2f}')
print(f'El costo de envío promedio de la tienda 2 es: ${envio_promedio2:.2f}')
print(f'El costo de envío promedio de la tienda 3 es: ${envio_promedio3:.2f}')
print(f'El costo de envío promedio de la tienda 4 es: ${envio_promedio4:.2f}')
```

```python


# Creo serie para los costos de envío promedio por tienda
costo_envio_promedio_por_tienda = pd.Series({
    'Tienda 1': envio_promedio,
    'Tienda 2': envio_promedio2,
    'Tienda 3': envio_promedio3,
    'Tienda 4': envio_promedio4
})


plt.figure(figsize=(10, 6))
costo_envio_promedio_por_tienda.plot(kind='line', marker='o', color=[store_colors[store] for store in costo_envio_promedio_por_tienda.index])
plt.title('Costo de Envío Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo Promedio de Envío ($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Costo de envío promedio**

El análisis del costo de envío promedio muestra que la Tienda 4 presenta el menor costo logístico entre las tiendas analizadas, seguida por la Tienda 3, Tienda 2 y Tienda 1.

Este resultado sugiere que el bajo desempeño comercial de la Tienda 4 no está relacionado con mayores costos de envío, ya que, pese a contar con costos logísticos más bajos, la tienda continúa registrando menores ingresos y menor volumen de ventas en comparación con las demás. Por lo tanto, **el costo de envío no constituye un factor determinante en su bajo rendimiento**, lo que refuerza la hipótesis de que existen otros factores comerciales o de demanda que afectan su desempeño.

##Conclusión global del análisis

A partir del análisis de los datos de ventas de las cuatro tiendas, considerando indicadores como ingresos totales, desempeño por categorías de productos, calificación promedio de clientes y costos de envío, se identifican diferencias relevantes en el rendimiento comercial de cada unidad.

Las **Tiendas 1 y 2** presentan el **mejor desempeño general**, evidenciando mayores ingresos totales, una mayor diversificación de categorías vendidas y una participación significativa en categorías de alto valor como electrónicos, electrodomésticos y muebles. Estos factores indican una operación comercial sólida y una mayor contribución al resultado global del negocio.

Por su parte, la **Tienda 3** muestra un **desempeño intermedio**, manteniendo niveles de ventas estables y una participación relativamente equilibrada en distintas categorías. Sin embargo, su volumen de ingresos es menor en comparación con las tiendas líderes.

En contraste, la **Tienda 4** presenta los indicadores de **desempeño más bajos** del conjunto analizado. Esta tienda registra menores ingresos totales, menor volumen de ventas, menor diversificación de categorías y un menor dinamismo comercial, lo que limita su contribución a la rentabilidad general de la empresa.
